```python
#C:\Users\BALASARATH\AppData\Local\Programs\Python\Python38

import cv2 
import os 
import numpy as np
from tkinter import ttk
from tkinter import filedialog
from docx.shared import Pt
from docx.enum.style import WD_STYLE_TYPE
from docx.enum.text import WD_LINE_SPACING
from docx2pdf import convert
from docx.enum.text import WD_BREAK
import docx, os, datetime
import pandas, random
from matplotlib import pyplot as plt
from docx.shared import Inches
from tkinter import *

from keras.models import load_model
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

video_file = "videos/test2_Trim.mp4"

emotion_dict = {}
#emotion_list = ["Angry", "Sad", "Happy", "Suprise", "Neutral", "Disgust", "Fear"]
emotion_list = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
emotion_data = []
word_doc = docx.Document()
empty_num = 0
# for i in range(0, 100):
# 	emotion_data.append(emotion_list[random.randint(0, 5)])

#print(emotion_data)

cropped_img_folder = ""
folder_name = ""

Name = "Arman Malik"
Id = "PSG01254"

def create_folder(name):
	#os.chdir(os.path.dirname(__file__))
	rand = 0
	try: 
		while True:
			folder_name = name + '(' + str(rand) + ')'
			if not os.path.exists(folder_name): 
				os.makedirs(folder_name)
				break
			else:
				rand += 1 
	except: 
		print ('Error while creating folder') 
	return folder_name

def extract_frames():
	global folder_name, video_file
	video = cv2.VideoCapture(video_file) 
	folder_name = create_folder("frames")
	frameno = 0

	while(True): 
		present,frame = video.read() 
		
		if present and frameno%5==0: 
			frame_name = './' + folder_name +'/frame' + str(frameno) + '.jpg'
			print ('Creating ' + str(frameno)) 
			cv2.imwrite(frame_name, frame) 
		elif not present: 
			break 
		frameno += 1
	video.release() 
	cv2.destroyAllWindows() 

def detect_crop_face():
    global folder_name, cropped_img_folder
    cropped_img_folder = create_folder("cropped_img")
    #image_with_box_folder = create_folder("boxed_img")

    model = cv2.dnn.readNetFromCaffe("models/deploy.prototxt", "models/weights.caffemodel")

    for file in os.listdir(folder_name):
        image = cv2.imread(folder_name + "/" + file)

        (h, w) = image.shape[:2]
        blob = cv2.dnn.blobFromImage(cv2.resize(image, (300, 300)), 1.0, (300, 300), (104.0, 177.0, 123.0))

        model.setInput(blob)
        detections = model.forward()

        #print(detections.shape)

        for i in range(0, detections.shape[2]):
            confidence = detections[0, 0, i, 2]
            if (confidence > 0.5):
                box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
                (startX, startY, endX, endY) = box.astype("int")
                cv2.rectangle(image, (startX, startY), (endX, endY), (255, 255, 255), 2)
                frame = image[startY:endY, startX:endX]

        #cv2.imwrite(image_with_box_folder + "/" + file, image)
        try:
            cv2.imwrite(cropped_img_folder + "/" +  file, frame)
            print("Image " + file + " converted successfully")
        except:
            continue


def second_header(patient_name, patient_id, link):
	abc = word_doc.add_paragraph("")
	abc.alignment = 0
	
	abc.add_run('PATIENT NAME:		 ', style='CommentsStyle')
	abc.add_run(patient_name + '\n', style='CommentsStyle').bold = False

	abc.add_run('PATIENT ID: 		 ', style='CommentsStyle')
	abc.add_run(patient_id + '\n', style='CommentsStyle').bold = False

	abc.add_run('LINK:', style='CommentsStyle')
	abc.add_run(link + '\n', style='CommentsStyle').bold = False

def first_header():
	#borders()
	abc = word_doc.add_paragraph("")
	abc.add_run('PSG INSTITUTE OF MEDICAL SCIENCE AND RESEARCH\n', style='CommentsStyle')
	abc.alignment = 1
	abc.add_run("PEELAMEDU, COIMBATORE - 641 004" + "\n", style='Secondstyle')
	abc.add_run("\nPATIENT EMOTION ASSESSMENT REPORT", style='CommentsStyle').underline = True

	second = word_doc.add_paragraph("")
	second.alignment = 2
	second.add_run('DATE: ' , style='Secondstyle').bold = True
	second.add_run(str(datetime.date.today().strftime("%d - %m - %Y")), style='Secondstyle')

def styles():
	obj_styles = word_doc.styles
	obj_charstyle = obj_styles.add_style('CommentsStyle', WD_STYLE_TYPE.CHARACTER)
	obj_font = obj_charstyle.font
	obj_font.size = Pt(14)
	obj_font.name = 'Calibri'
	obj_font.bold = True

	obj_charstyle1 = obj_styles.add_style('Secondstyle', WD_STYLE_TYPE.CHARACTER)
	obj_font1 = obj_charstyle1.font
	obj_font1.size = Pt(12)
	obj_font1.name = 'Calibri'

def convert_to_pdf(code):
	global empty_num
	if code:
		final_path = os.getcwd() + '\\' +  code + '.pdf'
	else:
		final_path = os.getcwd() + '\\' + '-----' + str(empty_num) + '.pdf'
		while os.path.exists(final_path): 
			final_path = os.getcwd() + '\\' + '-----' + str(empty_num+1) + '.pdf'
			empty_num += 1
	word_doc.save("Final Word1.docx")
	convert("Final Word1.docx", final_path)
	os.remove("Final Word1.docx")

def overall_analysis():
	abc = word_doc.add_paragraph("")
	abc.alignment = 0

	abc.add_run("OVERALL ANALYSIS: \n", style="CommentsStyle")
	tot_count = len(emotion_data)
	for i in emotion_data:
		emotion_dict[i] = emotion_dict.get(i, 0) + 1;
	for i in list(set(emotion_list) - set(emotion_data)):
		emotion_dict[i] = 0;
	for i in emotion_dict.keys():
		mod_data = i.upper()
		while 8 - len(mod_data) != 0: 
			mod_data += " " 
		 
		abc.add_run("	" + mod_data + " -   ", style="Secondstyle").bold = True
		abc.add_run(str(round((emotion_dict[i]/tot_count)*100, 2)) + "%\n", style="Secondstyle")

def frequency_graph():
	global empty_num, emotion_list
	fg_name = os.getcwd() + '\\' + 'fg_' + str(empty_num) + '.png'
	while os.path.exists(fg_name): 
		fg_name = os.getcwd() + '\\' + 'fg_' + str(empty_num+1) + '.png'
		empty_num += 1


	abc = word_doc.add_paragraph("")
	abc.alignment = 0 
	abc.add_run("DETAILED ANALYSIS(FREQUENCY GRAPH): \n\n", style="CommentsStyle")

	#plt.axvline(label='Frequency of emotion', linewidth=2)
	#plt.legend(loc ="upper right") 
	# plt.hist(emotion_data, bins=emotion_list, edgecolor="white", color='#607c8e', rwidth=0.9, align='mid')

	fig = plt.figure()
	
	plt.bar(list(emotion_dict.keys()), list(emotion_dict.values()), color='#607c8e', width=0.4)
	plt.title("Emotion of patients")
	plt.xlabel("Emotion")
	plt.ylabel("Frequency")
	plt.style.use('bmh')
	#plt.tight_layout()
	#plt.show()
	plt.savefig(fg_name, bbox_inches = 'tight')
	abc.add_run().add_picture(fg_name, width=Inches(6), height=Inches(3.4))
	os.remove(fg_name)
	#plt.show()
	plt.clf()

def time_emotion_graph():
	global empty_num
	teg_name = os.getcwd() + '\\' + 'teg_' + str(empty_num) + '.png'
	while os.path.exists(teg_name): 
		teg_name = os.getcwd() + '\\' + 'teg_' + str(empty_num+1) + '.png'
		empty_num += 1


	abc = word_doc.add_paragraph("")
	abc.alignment = 0 
	abc.add_run("DETAILED ANALYSIS(TIME EMOTION GRAPH): \n\n", style="CommentsStyle")

	plt.style.use('ggplot')
	x = [i for i in range(0, len(emotion_data))]
	y = emotion_data
	plt.plot(x, y)
	plt.xlabel("Time")
	plt.ylabel("Emotion")
	plt.title("Emotion at specific time interval")
	#plt.show()
	plt.savefig(teg_name, bbox_inches = 'tight')
	abc.add_run().add_picture(teg_name, width=Inches(6), height=Inches(3.7))
	os.remove(teg_name)
	plt.clf()


def disclamer():
	abc = word_doc.add_paragraph("")
	abc.alignment = 0

	abc.add_run("\nDISCLAMER: ", style="Secondstyle").bold = True
	abc.add_run("This is a machine generated report. Please verify the original data for the correctness of the report.", style="Secondstyle")

def post_processing():
    global Name, Id, cropped_img_folder
    styles()
    first_header()
    second_header(Name, Id, os.getcwd() + "\\" + cropped_img_folder)
    overall_analysis()
    frequency_graph()
    time_emotion_graph()
    disclamer()
    convert_to_pdf("")

def emotion_analysis(emotions):
    objects = ('Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise')
    #bjects = ("Angry", "Sad", "Happy", "Suprise", "Neutral", "Disgust", "Fear")
    #["Angry", "Sad", "Happy", "Suprise", "Neutral", "Disgust", "Fear"]
    y_pos = np.arange(len(objects))
    
    plt.bar(y_pos, emotions, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.ylabel('percentage')
    plt.title('emotion')
    
    plt.show()

def run_model(file, model):
    
    true_image = image.load_img(file)
    file = np.array(true_image)
    cropped_img=cv2.resize(np.float32(file),(96,96))

    x = image.img_to_array(cropped_img)
    x = np.expand_dims(x, axis = 1)

    emotions = {0: "Angry", 1: "Disgust", 2: "Fear", 3: "Happy", 4: "Neutral", 5: "Sad", 6: "Surprise"}
    x /= 255
    x = np.array(x, 'float32')
    x = x.reshape(-1, 96, 96, 3)
    predictions = model.predict(x)
    print(predictions)
    prediction = np.argmax(predictions)
    result = emotions[prediction]
    print(result)
    emotion_data.append(result)
    emotion_analysis(predictions[0])
    
    plt.gray()
    plt.imshow(true_image)
    plt.show()
    
    
def implementation():
    model = load_model('models/Xception.h5')
    global cropped_img_folder
    for i in os.listdir(cropped_img_folder):
        run_model(cropped_img_folder+"//"+i, model)
    global emotion_data
    print(emotion_data)


def main():
	#pre-processing
	extract_frames()
	detect_crop_face()

	# #implementation
	implementation()

	# #report generation
	post_processing()

main()
```
